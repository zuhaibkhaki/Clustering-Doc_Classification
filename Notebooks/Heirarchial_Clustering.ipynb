### **Heirarchial Clustering:-**

There are two types of heirarchial clustering:-

**i)** Agglomerative Clustering (Mostly Used)

**ii)** Divisive Clustering


**Agglomerative Clustering:-**

**Step 01:-** Initialize the proximity matrix.

Matrix representing the distance beteen clusters is the proximity matrix. If there are n data points in our scatter plot, then the proximity matrix is a square symmetric matrix of order n*n.

**Step 02:-** Make a cluster around each point.

**Step 03:-** Inside a loop:

  **i)** Maerge the closest 02 clusters.

  **ii)** Update the proximity matrix.

**Step 04:-** Continue this until only one cluster is left.

**Divisive Clustering:-**

Divisive clustering works the same way as agglomerative clustering but just the process is reversed ie. it starts from considering the entire dataset as a single cluster, and then starts to divide the single cluster into sub-clusters untill we've those many clusters as the no. of data points.

There are different types of agglomerative clustering based on how the distance between two clusters is calculated:-



**i)** **Single-Link (Min):-**

  In Single-Link agglomerative clustering, the intercluster distance (intercluster similarity) is calculated by taking the distance between two nearest points.

id1.JPG

**ii)** **Complete Link (Max):-**

  In Complete Link agglomerative clustering, the intercluster distance (intercluster similarity) is calculated by taking the distance between two farthest points.

id2.JPG

**iii)** **Group Average:-**

  In Group Average agglomerative clustering, the intercluster distance (intercluster similarity) is calculated by taking the average of the distance between all combination of points.

id3.JPG

**iv)** **Ward:-**

  In Ward, the intercluster distance is calculated as denoted in the image.

###**Finding the Optimum No. Of Clusters:-**

In case of heirarchial clustering, the intercluster distance is calculated by plotting a dendogram.

In the dendogram, we've to look for the longest vertical line which is not cut by any horizontal line. When we locate such a vertical line, we draw a horizontal line cutting that vertical line from middle and find the no. of points at which the horizontal line cuts the vertical line. These no. of intersection points represent the optimum no. of clusters.

Dendogram.JPG

##**Hyperparameters:-**

**i)** **n_clusters:-**

It specifies the number of clusters in the scatter plot

**ii)** **affinity:-**

It specifies the type of distance to be taken between the clusters and the default value is *euclidean*

**iii)** **linkage:-**

It specifies the type of agglomerative clustering. It can take following values:
*single,complete,average,ward*. The default value is *ward*

**iv)** **distance_threshold:-**

It specifies the minimum distance below which the clustering won't take place.

**Benefits:-**

**i)** Heirarchial Clustering is widely applicable unlike K-Means where the data was required to be spherical and well seperated from eachother. But, since heirarchial clustering has different types, if one fails to cluster the points, other type can be implemented.

**ii)** Due to presence of a dendogram, we're able to record the history of clustering ie. which point is getting clustered to which point.

**Limitations:-**

**i)** If we've n data points in our scatter plot, out proximity matrix has to contain n square points which consumes a lot of memory. So, it can't be useful for large datasets

###**Code Implementation:-**

```python
#Importing the required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Loading the required dataset
data=pd.read_csv("student_clustering.csv")
data.head()
```

```python
#Generating the dendogram
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = sch.dendrogram(sch.linkage(data, method='ward'))
```

From the dendogram, it's quite evident that the optimum no. of clusters are 02

```python
#Heirarchial clustering

#Importing the required library
from sklearn.cluster import AgglomerativeClustering
```

```python
#Creating an object of AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
```

```python
#Fitting the model into data
clustering.fit(data)
```

```python
#Extracting the labels
labels=clustering.labels_
```

```python
#Making a scatter plot
plt.scatter(data["cgpa"],data["iq"],c=labels)
plt.show()
```
