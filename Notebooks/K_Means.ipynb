###**K-Means:-**

It is a **centroid based** clustering algorithm.

**Steps:**

**1.** Decide n Clusters.

  You've to tell the algorithm how many clusters you want your data to be clustered into. This is infact one of the limitations of K-Means.

**2.** Initialize Centroids.

  Based on how many clusters your data has, those many centroids will be initialized by our algorithm randomly.

**3.** Based on these centroids, we assign a cluster to each point.

  This is done by taking the **Euclidean distance** of each data point from these 03 centroids. Feom whichever, centroid this distance comes out to be **minimum**, that data point is assigned a cluster of that centroid.

**4.** Move centroids.

  Now since we've clustered our data. We find the centroids of each of the cluster by taking the mean of x values and y values of data points of that cluster.

**5.** Assign cluster to each point on the basis of these new centroids.

  Based on these new centroids, we assign clsuters to each data point again.

**6.** Move centroid.

  Then we again try to move the centroid by taking the mean of x values and y values of cluster points of a particular cluster. If the centroids move from their exisiting position, we assign clusters to points again on the basis of new centroids.

**7.** The process stops, when new centroids coincide with the previous centroids and then we can say that we've acheived the best possible clustering.

**Now, how do we determine the optimum no. of clusters we should take ?**

For this, we use the **Elbow Method**

Elobow method uses a curve called, **Elbow Curve** to locate an **Elbow Point** which denotes the optimum number of clusters we should take in our clustering.

Elbow curve is a curve which has something called WCSS (Within Cluster Sum of Squared Distance) on its y-axis and the No. of clusters on its x-axis.

The within cluster sum of squared distances (WCSS) is the sum of the squared distance between each point and the centroid in a cluster.

So, for a single cluster, WCSS is simply the sum of distances of points from the centroid and for multiple clusters, WCSS is the sum of WCSS of individual clusters.

As, the number of clusters increase, the WCSS decreases.
This can be proved logically. If we've a single cluster in our data, so the WCSS would have some value. Now let's say our data contains 50 data points. So, the maximum no. of clusters we can have is 50 ie. each point behvaes as a cluster of its own. So in that case, the WCSS would be zero.

Basically, as we move from cluster 1 to 2, our WCSS decreases significanltly. This means on going from one to two clusters, we've improved our results. Similarly on going from 02 to 05 clusters a sharp decrease is found. But now, as we go from 05 clusters to 06, nothing significant happens. So, we say 05 was our optimum no. of clusters and it also happens to be the elbow point.

```python
#Loading a dummy dataset
data=pd.read_csv("student_clustering.csv")
data.head()
```

**Code Implementation:-**

* **How to generate elbow curve:-**

```python
#Importing the required libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#Create an empty list to store the within cluster sum of squared distance (WCSS) values:
wcss=[]
```

```python
#Iterate through a range of cluster numbers
for i in range(1,11):
  #Create a KMeans object with the current cluster number
  kmeans=KMeans(n_clusters=i)
  #Fit your data into the model
  kmeans.fit(data)
  #Append the wcss value to the list
  wcss.append(kmeans.inertia_)
```

```python
#Plot the wcss values against the cluster numbers
plt.plot(range(1,11),wcss)
plt.show()
```

```python
#Choose the optimal number of clusters
#The elbow curve will typically have a sharp bend at the optimal no. of clusters. Choose the npo. of clusters where the curve starts to flatten out.
```

* **How to apply K-Means on dataset:-**

```python
#Import the necessary libraries
from sklearn.cluster import KMeans
```

```python
#Choose the no. of clusters
num_clusters=3
```

```python
#Initialize the KMeans model
kmeans=KMeans(n_clusters=num_clusters)
```

```python
#Fit the model on your data
kmeans.fit(data)
```

```python
#Get the cluster labels
labels=kmeans.labels_
labels
```

```python
#Get the centroids
centers=kmeans.cluster_centers_
centers
```

```python
#Visualize the clustered data
plt.scatter(data["cgpa"],data["iq"],c=labels)
plt.scatter(centers[:,0],centers[:,1],marker="*",c="red",s=100)
plt.show()
```
