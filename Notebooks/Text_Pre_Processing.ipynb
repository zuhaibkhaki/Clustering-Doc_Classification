**Removing HTML Tags:-**

HTML tags are used to instruct the browser, how to display the information on the webpage. These are found in abundance in data extracted from webpages/APIs and thus need treatment.

These are removed using the concept of *regular expressions (regex)*

````python
#Generating a test string
str2="<p> Welcome To Our Page </p>"
````

````python
#Removing HTML tags

#Importing the re module
import re

#Creating a function to remove HTML tags
def remove_html_tags(text):
  pattern=re.compile("<.*?>")
  return(pattern.sub(r"",text))
````

````python
#Invoking the function
remove_html_tags(str2)
````

**Removing URLs:-**

Often our data contains URLs in the text which are meaningless for our machine learning algorithm. So, we need to remove such URLs and the way we do it is again using *regular expressions (regex)*

````python
#Generating a test string
str3="You can check my profile at https://www.facebook.com/"
````

````python
#Removing URLs

#Importing the re module
import re

#Crearting a function to remove URLs
def remove_url(text):
  pattern=re.compile(r"https?://\s+|www.\.\s+")
  return(pattern.sub(r" ",text))
````

````python
#Invoking the function
remove_url(str3)
````

**Removing Punctuation Marks:-**

Punctuation Marks are symbols/characters which are used in english language construction, to improve readability and comprehension.

It's important to remove these symbols and this can be understood by following argument:

  Let's say we've a string "Hello!,how are you?".When we tokenize the string, there are two possibilities:

  **i)** ["Hello","!","how","are","you","?"] ie. Either punctuation marks are considered as seperate words.

  This approach is problematic since punctuation marks carry no meaning individually and they act just like the noise in the list. They increase the complexity and nothing else.

  **ii)**["Hello!","how","are","you?"] ie. Or, punctuation marks are taken along with their preceding words.

  This approach is problematic because python will interpret "Hello!" and "Hello" as two different entitites whenever it encounters "Hello" in other parts of string.
  
So, it is necessary to remove punctuation marks.

To access all the punctuation marks in python we just have to write following lines of code,



```
import string
string.punctuation
```



````python
#Importing the string module
import string
````

````python
#Getting the list of all punctuation marks
punc=string.punctuation
punc
````

````python
#Generating a test string
str4="Hey,how are you doing?"
````

````python
#Creating a user-defined function
def remove_punctuation(text):
  for i in punc:
    text=text.replace(i," ")
  return(text)
````

````python
#Function invoking
remove_punctuation(str4)
````

**Removing StopWords:-**

Stopwords are words like "of","the","a","an" which are used for sentence construction but they do not carry an individual meaning. They're basically conjunctions and articles and need to be removed.

**NLTK** library provided a list of such stopwords from different languages and we use it to remove stopwords from our data.

````python
#Importing required library
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")
````

````python
#Displaying the stopwords
stopwords.words("english")
````

````python
#Generating a test string
str5="I am going to delhi and assam"
````

````python
#Defining a function to remove stopwords
def remove_stopwords(text):
  new_txt=[]
  for i in text.split(" "):
    if i in stopwords.words("english"):
      new_txt.append(" ")
    else:
      new_txt.append(i)
  return(" ".join(new_txt))
````

````python
#Invoking our function
remove_stopwords(str5)
````

**Handling Emojis:-**

Emojis are digital miniature emojis which are used to represent emotions. So there are two ways to handle them since they can't be understood by the ML algorithm.

````python
#Creating a dummy string
str6="Python is 🔥"
````

````python
#Method 01: Removing the emojis

#Import the RE module
import re

#Define the regular expression pattern
emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # Emoticons
    u"\U0001F300-\U0001F5FF"  # Miscellaneous Symbols and Pictographs
    u"\U0001F680-\U0001F6FF"  # Transport and Map Symbols
    u"\U0001F1E0-\U0001F1FF"  # Flags (iOS)
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
"]+", flags=re.UNICODE)

#Use the sub method to replace emojis with an empty string
def remove_emoji(text):
    return emoji_pattern.sub(r'', text)

#This function takes a string as input and returns a new string with all emojis removed.
````

**Tokenization:-**

Tokenization is the technique of splitting our data/strings into small parts called "tokens".These tokens can be phrases, setences, words etc. This step is most important since if we're able to generate perfect tokens for our data, our embeddings are also going to be flawless.

There are two types of tokenization:

**i)** **Word Tokenization:**

In this type of tokenization, every word of the string gets converted into a token. Eg. "I am going to Delhi" will get tokenized into ["I","am","going","to","Delhi"]

**ii)** **Sentence Tokenization:**

In this type of tokenization, every sentence of the string gets converted into a token. Eg. "I am going to Delhi.I am on holiday" will get tokenized into ["I am going to Delhi","I am on holiday"]

Some of the techniques of tokenization are as follows:

**Stemming:-**

In grammer, **Inflection** is the modification of a word to express different grammatical categories such as a tense, case, voice, person, number etc.

Eg. The word "walk" can be changed into "walking","walks","walked" by adding different suffixes.
So, the category of root word/stem is changed and this is called inflection.

**Stemming** is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not meaningful.

The algorithms which enable us to do so are called **stemmers**. These algorithms are developed by linguists who are experts of a language. They are mostly used in **information retrevial systems** as in google.Eg. In google if you search "fish" or "fishing" or "fish net", you will get somewhat similar results.

The stemmers are available in **NLTK** library and the stemmer for **english language** is called **porter stemmer**. For other languages we use **snowball stemmer**.

````python
#Importing the required library
from nltk.stem.porter import PorterStemmer
````

````python
#Creating an object for the stemmer
ps=PorterStemmer()
````

````python
#Creating a function for stemming
def stemming(text):
  new_text=[]
  for i in text.split():
    new_text.append(ps.stem(i))
  return(" ".join(new_text))
````

````python
#Creating a test string
str7="I am comming probably"
````

````python
#Invoking the function
stemming(str7)
````

The problem with stemming is that it reduces the inflection to bring a word back to its stem/root word, but the stem may not be meaningful always.
So, if you've to display your end result to a user, you can't display the stemmed data.

For displaying the final output to user, you have to use Lemmitization. But, lemmatization is slower than stemming.

Both lemmatization and stemming do the same thing but the result of stemming ie. stem may/may not be a valid word. Whileas, the output of lemmatization is a valid word in english.

So, if the output is not to be shown to the user and speed matters to us, we use stemming but if we want to display output to user, we use lemmatization.

**Lemmatization:-**

Lemmatization, unlike stemming, reduces the inflected words properly ensuring that the root words belongs to english language.

In lemmatization, the root word is called **lemma**

In lemmatization, we go to a dictionary to map the stem to the lemma. Here, we use **wordnet lemmatizer**.

So, in stemming there are algorithms that remove the suffixes and prefixes but in lemmatization, we search for the lemma and thus it is slow.

````python
#Importing the required library
import nltk
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
````

````python
#Creating an object for lemmatizer
wordnet_lemmatizer=WordNetLemmatizer()
````

````python
#Greating a test string"
str10="I am running"
````

````python
#Defining a function for lemmatization
def lemmatization(text):
  new_text=[]
  for i in text.split():
    new_text.append(wordnet_lemmatizer.lemmatize(i,pos="v"))
  return(" ".join(new_text))
````

````python
#Invoking the function
lemmatization(str10)
````
