###**Silhouette Score:-**

Silhouette score is the method of checking the accurary of our clustering algoritm.

This score basicslly refers to the meausre of how similar object is to its own cluster (cohesion) compared to other clusters (seperation).

The value of this coefficient ranges from **-1 to +1**, where a high value indicats that the object is well matched to its own cluster and poorly matched to neighbouring clusters.

The silhouette value can be calculated with any distance metrics such as Euclidean distance or the Manhattan distance.

**Mathematics behind silhouette score:-**

Assume that we have clustered our data using any clustering algoritm.

We take a cluster and choose a point within it. Then within that cluster, we take the distance of every point from our current point ie. intercluster distance of points from our chosen point. Take the avergae of all these distances.

We then define the mean dissimilarity of points i to some cluster Ck as the mean of the distance from i to all points in Ck(where Ck != Ci)

```python
#Importing the required dataset
data=pd.read_csv("student_clustering.csv")
data.head()
```

```python
#Import the necessary libraries
import pandas as pd
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
#Clustering the data using K-Means
kmeans=KMeans(n_clusters=3)
kmeans.fit(data)
labels=kmeans.labels_
```

```python
#Calculate the silhuette score
score=silhouette_score(data,labels)
#data: A 2D array representing data points
#labels: A 1D array representing cluster labels
print(score)
```

```python
#Interpret the score
#The silhouette score ranges from -1 to +1, where:
# -1: indiciates that the data points is assigned to the wrong cluster
# 0: indicates that the data point is on the border of two clusters
# +1: indicates that the data point is well clustered
```
